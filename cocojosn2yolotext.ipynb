```python

import json
import os
cnt = 0

with open('/content/drive/MyDrive/Data Analyst/Kwater/train.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)
    iw = json_data['images'][0]['width']  # 원본 이미지 넓이
    ih = json_data['images'][0]['height'] # 원본 이미지 높이
    json_f = {j['id']: j['file_name'] for j in json_data['images']}
    print(json_f)
    annotations =  json_data['annotations']
    for annot in annotations:
        file = open('/content/drive/MyDrive/Data Analyst/Kwater/labels/'+json_f[annot['image_id']].split('.')[0]+'.txt', 'w')
        bx, by, bw, bh = annot['bbox'] # 객체 1개의 bbox 위치 좌표
        x = (bx + ( bw / 2 )) / iw # x 좌표
        y = (by + ( bh / 2 )) / ih # y 좌표
        w = bw / iw # w 좌표
        h = bh / ih # h 좌표
        cls = annot['category_id']
        file.write(f"{cls} {x:f} {y:f} {w:f} {h:f}\n")
        file.close()
        cnt += 1

print(cnt)
```
